#About this notebook

Here, we run Logistic Regression on the interday stocks we had identified to be part of the causal chain for Apple, and we train the model and test it to see what kind of accuracy we get (50-55%)

```
from sklearn.linear_model import LogisticRegression
import pandas_datareader as pdr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
stock_list=['AAPL','CHL','WMT','PG','GPC','PEP']
master_df=pd.DataFrame()
data_AAPL=pdr.DataReader('AAPL',data_source='yahoo', start='23/4/2019', end='10/6/2020')
master_df['AAPL_price']=data_AAPL['Close']
for stock in stock_list:
  data=pdr.DataReader(stock, data_source='yahoo', start='23/4/2019', end='10/6/2020')
  percentage_change=data['Close'].pct_change().to_list()
  signal=[]
  for x in percentage_change:
    if (x>=0.0):
      signal.append(1)
    elif (x<0.0):
      signal.append(0)
    else:
      signal.append(-1)
  master_df['signal_'+str(stock)]=signal
```

```
master_df
```

```
sCHL=master_df['signal_CHL'].to_list()
sGPC=master_df['signal_GPC'].to_list()
sPEP=master_df['signal_PEP'].to_list()
sPG=master_df['signal_PG'].to_list()
sWMT=master_df['signal_WMT'].to_list()
sAAPL=master_df['signal_AAPL'].to_list()
s_aapl=sAAPL[15:195]
s_chl=sCHL[13:193]
s_pep=sPEP[0:180]
s_pg=sPG[9:189]
s_gpc1=sGPC[1:181]
s_gpc2=sGPC[2:182]
s_wmt=sWMT[11:191]
a=list(zip(s_chl,s_wmt,s_pg,s_pep,s_gpc1,s_gpc2))
X=np.array(a)
```

```
clf=LogisticRegression(random_state=5).fit(X,s_aapl)
```

```
clf
```

```
CHL123=sCHL[193:263]
PEP123=sPEP[180:250]
PG123=sPG[189:259]
GPC1231=sGPC[181:251]
GPC1232=sGPC[182:252]
WMT123=sWMT[191:261]
b=list(zip(CHL123, PEP123, PG123, GPC1231, GPC1232, WMT123))
X1=np.array(b)
```

```
X_pred=clf.predict(X1)
```
